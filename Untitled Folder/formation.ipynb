Visualisation

```python
#histogramme
plt.hist(df_cleaned['Data_value'], bins=5,color='skyblue',edgecolor='black')
plt.xlabel("valeurs")
plt.ylabel("Frequence")
plt.title("Test")
plt.show()
```

```python
# en barre
plt.bar(df_cleaned['Period'],df_cleaned['Data_value'], color=['red','blue','green','purple'])
plt.xlabel("Periode")
plt.ylabel("Valeur par groupe")
plt.show()
```

```python
# nuage de point
plt.scatter(df_cleaned['Period'],df_cleaned['Data_value'], color='blue',alpha=0.5)
plt.xlabel("Periode")
plt.ylabel("Valeur par groupe")
plt.title('nuage de point')
plt.show()
```

```python
# courbe ligne
plt.plot(df_cleaned['Period'],df_cleaned['Data_value'],marker='o',linestyle='-',color='blue')
plt.xlabel("Periode")
plt.ylabel("Valeur par groupe")
plt.title('Courbe ligne')
plt.show()
```

```python
#seaborn(histogramme ameliorée)
import seaborn as sns
```

```python
sns.histplot(df_cleaned['Data_value'],bins=5,kde=True,color='blue')
plt.title("Distibution")
plt.show()
```

```python
# boxlot
sns.boxplot(x=df_cleaned['Data_value'],color='orange')
plt.title("Boxplot")
plt.show()
```

```python
sns.barplot(x='Period', y='Data_value', data=df_cleaned, palette='viridis')
plt.title("Graphique en barres avec Seaborn")
plt.show()
```

```python
sns.regplot(x='Period', y='Data_value', data=df_cleaned)
plt.title("Nuage de points avec régression")
plt.show()
```
