```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['Sex']=encoder.fit_transform(df['Sex'])
```

```python

df['Survived'].value_counts().plot(kind='bar')
plt.title('Répartition des suirvivants par sexe')
plt.show()
```

```python
#Répartion par classe
df['Pclass'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Classe des passagers')
plt.show()
```

```python
#taux de survi par class
df.groupby('Pclass')['Survived'].mean()
```

```python
#taux de survi par sex
print(df.groupby('Sex')['Survived'].mean())
```

```python
#visualisation de taux de survie par classe et sexe
import seaborn as sns
sns.barplot(x='Pclass',y='Survived',hue='Sex',data=df)
plt.title('Taux de survie  par classe et sexe')
plt.show()
```

```python
df_numeric = df.select_dtypes(include=['int64','float64'])
```

```python
corr = df_numeric.corr()
```

```python
print(corr)
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm',linewidths=0.5)
plt.title("Coorelation")
plt.show()
```
