```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
```

```python
#ajout des colonnes pour faire la prediction
titanic = df[['Survived','Pclass','Sex','Age']]
```

```python
titanic
```

```python
y = titanic['Survived']
x=titanic.drop('Survived',axis=1)
```

```python
#entrainnement de la modèle
model.fit(x,y)
model.score(x,y)
```

```python
#focntion pour la prediction
def survie(model,Pclass=2,Sex=1,Age=12):
    x = np.array([Pclass,Sex,Age]).reshape(1,3)
    res = ["Passager vivant" if model.predict(x)==1 else "Passager mort"]
    return res
```

```python
survie(model)
```
