```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([1.8, 3.9, 5.8, 8.2, 10.1])
```

```python
#création d'un modl de regression linéaire
model = LinearRegression()
```

```python
#entrainnement du modele sur x,y
model.fit(X,y)
```

```python
#prediction des valeurs de y à partir de X 
y_pred = model.predict(X)
```

```python
# Calcul de l'erreur quadratique moyenne (MSE) entre les vraies valeurs et les prédictions
mse = mean_squared_error(y, y_pred)
```

```python
# Affichage de l'erreur quadratique moyenne, arrondie à deux décimales
print(f"Erreur quadratique moyenne : {mse:.2f}")
```

```python
plt.scatter(X,y,color='blue',label='Données')
plt.plot(X, y_pred, color='red', label='Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regression linéaire simple')
plt.legend()
plt.show()
```
