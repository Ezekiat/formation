```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../Mall_Customers.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler,LabelEncoder
encoder = LabelEncoder()
```

```python
df['Genre'] = encoder.fit_transform(df['Genre'])
```

```python
df['Income'] = df['Annual Income (k$)']
df
```

```python
scaler = StandardScaler()
X_scaler = scaler.fit_transform(df[['Age','Income']])
```

```python
kmeans = KMeans(n_clusters=3,random_state=42)
df['cluster']= kmeans.fit_predict(X_scaler)
df
```

```python
df.head(50)
```

```python
import seaborn as sns
```

```python
sns.scatterplot(data=df, x='Income', y='Age', hue='cluster', palette='Set2')
plt.title("Clustering des clients")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Age")
plt.grid(True)
plt.show()
```
