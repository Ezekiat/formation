```python
# sÃ©paration de features
X = df.drop('Outcome',axis=1)
y = df['Outcome']
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
#model
model = LogisticRegression()
```

```python
#separation des donnÃ©es
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay,classification_report,accuracy_score
y_pred = model.predict(X_test)
```

```python
print("confusiopn matrix: ",confusion_matrix(y_test,y_pred))
print("classification report: ", classification_report(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test,y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Sain", "Malade"])
disp.plot(cmap='Blues')
plt.title("Matrice de confusion")
plt.show()
```

```python
def is_diabet(model, Pregnancies=0,Glucose=93,BloodPressure=40,SkinThickness=3,Insulin=0,BMI=33.6,DiabetesPedigreeFunction=0.627,Age=50):
    x = np.array([Pregnancies,Glucose,BloodPressure,SkinThickness,Insulin,BMI,DiabetesPedigreeFunction,Age]).reshape(1,-1)
    res = ["diabÃ¨tique" if model.predict(x)==1 else "sain"]
    proba = model.predict_proba(x)[0][1]
    print(f"Resultat: {res}, avec la probabilitÃ© {round(proba*100,2)}%")
```

```python
is_diabet(model)
```
